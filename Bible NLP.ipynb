# NLP'ing the Bible

## Acquiring the Sacred Texts

```python
import requests
from bs4 import BeautifulSoup
import urllib.request, urllib.parse, urllib.error

bibleUrl = 'http://www.stewartonbibleschool.org/bible/text/index.html'
baseUrl = 'http://www.stewartonbibleschool.org/bible/text/'
#entireBible = 'http://www.gutenberg.org/files/10/10-0.txt'
page = requests.get(bibleUrl)
soup = BeautifulSoup(page.content, 'html.parser')
links = soup.find_all('a')
#print(links)
bookUrls = [l.get('href',2) for l in links]
bookUrls = [b for b in bookUrls if '.txt' in b]
print(bookUrls)

#baseUrl + bookUrls[0]
books = {}
for url in bookUrls:
    books[url] = urllib.request.urlopen(baseUrl + url).read().decode('utf-8') # matthew is 39?
```

## Summarizing Text

```python
import re
# https://regex101.com/
bookText = re.sub('[0-9]*:[0-9]*:','',bookText)
print(bookText)
```

```python
# https://github.com/DivakarPM/NLP/blob/master/Text_Summarization/Text_Summarization.ipynb
import nltk
from string import punctuation
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize

nltk.download('punkt')
nltk.download("stopwords")
stop_words = stopwords.words('english') + ['thee','thou','thy','thine','ye','unto','shall']
punctuation = punctuation + '\n'

bookText = '\n'.join(books.values()) # THE WHOLE BIBLE
tokens = word_tokenize(bookText)
# We need to count the occurances of each non-stopword in the text
word_frequencies = {}
for word in tokens:    
    if word.lower() not in stop_words:
        if word.lower() not in punctuation:
            if word not in word_frequencies.keys():
                word_frequencies[word] = 1
            else:
                word_frequencies[word] += 1

#print(word_frequencies)

# We want to weight the frequency by the most common word
max_frequency = max(word_frequencies.values())
#print(max_frequency)
for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word]/max_frequency
print({k:v for k, v in sorted(word_frequencies.items(), key=lambda item: item[1],reverse=True)})
```

```python
# We can score each sentence by summing up the weighted frequency of each word in the sentence
# We're looking for the sentences that are at the "core" of frequent words/topics
from collections import OrderedDict

bookText = bookText.replace(';\n','.\n') # the geneology goes on forever, lets break it up
sent_token = sent_tokenize(bookText)
sentence_scores = OrderedDict() #{}
for sent in sent_token:
    sentence = sent.split(" ")
    for word in sentence:        
        if word.lower() in word_frequencies.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent] = word_frequencies[word.lower()]
            else:
                sentence_scores[sent] += word_frequencies[word.lower()]
print(sentence_scores)
```

```python
from heapq import nlargest
select_length = int(len(sent_token)*0.01)
print(select_length)

threshold = sorted(sentence_scores.values(),reverse=True)[select_length-1]

n_largest = {k:v for k,v in sentence_scores.items() if v >= threshold}
my_summary = [s for s in n_largest.keys()]
print(my_summary)

#summary = nlargest(select_length, sentence_scores, key = sentence_scores.get)
#print(summary)
```

## Topic Analysis

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

vectorizer = CountVectorizer(stop_words = stop_words,min_df=2)
tfidf = TfidfTransformer(smooth_idf=False)

bow = vectorizer.fit_transform(books.values()) #sent_token
print(bow.shape)
tfidf_bow = tfidf.fit_transform(bow.toarray())
vocabulary = vectorizer.get_feature_names()

print(len(vectorizer.get_feature_names()))
```

```python
from sklearn.decomposition import LatentDirichletAllocation
import pyLDAvis.sklearn

lda = LatentDirichletAllocation(n_components=100, random_state=0)
topics = lda.fit_transform(bow)
print(topics)

pyLDAvis.enable_notebook()
pyLDAvis.sklearn.prepare(lda,bow,vectorizer,sort_topics=False)
```
